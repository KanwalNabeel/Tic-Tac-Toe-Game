```python
Player1Win = 0
Player2Win = 0
Player1Loss = 0
Player2Loss = 0
DrawCount = 0
print("Tic-Tac-Toe Game \n\n")

while True:    
    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']    
    player = 1      
    Win = 1    
    Draw = -1    
    Running = 0        
    Game = Running 
    
    
    #print("Tic-Tac-Toe Game \n\n")   
    print('Player1 Choose the sign X or O')
    ch1 = input()
    print('Player2 Choose the sign X or O')
    ch2 =input()
    print('%s  : Player1  , %s : Player2 ' % (ch1 , ch2))
    
    def DrawBoard2():    
        print(" %c | %c | %c " % (board[1],board[2],board[3]))    
        print("_1_|_2_|_3_")    
        print(" %c | %c | %c " % (board[4],board[5],board[6]))    
        print("_4_|_5_|_6_")    
        print(" %c | %c | %c " % (board[7],board[8],board[9]))    
        print(" 7 | 8 | 9 ")
        
    print('\nPostions for Input')
    DrawBoard2()
   
    #This Function Draws Game Board    
    def DrawBoard():    
        print(" %c | %c | %c " % (board[1],board[2],board[3]))    
        print("___|___|___")    
        print(" %c | %c | %c " % (board[4],board[5],board[6]))    
        print("___|___|___")    
        print(" %c | %c | %c " % (board[7],board[8],board[9]))    
        print("   |   |   ")    
   
    #This Function Checks position is empty or not    
    def CheckPosition(x):    
        if(board[x] == ' '):    
            return True    
        else:    
             return False    
   
    #This Function Checks player has won or not    
    def CheckWin():    
        global Game    
    #Horizontal winning condition    
        if(board[1] == board[2] and board[2] == board[3] and board[1] != ' '):    
            Game = Win    
        elif(board[4] == board[5] and board[5] == board[6] and board[4] != ' '):    
            Game = Win    
        elif(board[7] == board[8] and board[8] == board[9] and board[7] != ' '):    
            Game = Win    
    #Vertical Winning Condition    
        elif(board[1] == board[4] and board[4] == board[7] and board[1] != ' '):    
            Game = Win    
        elif(board[2] == board[5] and board[5] == board[8] and board[2] != ' '):    
            Game = Win    
        elif(board[3] == board[6] and board[6] == board[9] and board[3] != ' '):    
            Game=Win    
    #Diagonal Winning Condition    
        elif(board[1] == board[5] and board[5] == board[9] and board[5] != ' '):    
            Game = Win    
        elif(board[3] == board[5] and board[5] == board[7] and board[5] != ' '):    
            Game=Win    
    #Match Tie or Draw Condition    
        elif(board[1]!=' ' and board[2]!=' ' and board[3]!=' ' and board[4]!=' ' and board[5]!=' ' and board[6]!=' ' and board[7]!=' ' and board[8]!=' ' and board[9]!=' '):    
             Game=Draw    
        else:            
             Game=Running    
    
    
    while(Game == Running):   
        if(player % 2 != 0):    
            print("\nPlayer 1's Turn")    
            Mark = ch1 
        else:    
            print("\nPlayer 2's Turn")    
            Mark = ch2    
        choice = int(input("Select Your Move "))    
        if(CheckPosition(choice)):
            board[choice] = Mark 
            DrawBoard()
            player+=1    
            CheckWin()    
     
       
    if(Game==Draw):
        DrawCount += 1
        print("Game Draw")    
    elif(Game==Win):    
        player-=1    
        if(player%2!=0):
            Player1Win += 1
            Player2Loss += 1
            print("\n----Player 1 Won----\n") 
        else:    
            Player2Win += 1
            Player1Loss += 1
            print("\n----Player 2 Won----\n")
            

    print('Player1--Wins & Losses:')
    print('%s Wins, %s Losses, ' % (Player1Win, Player1Loss))
    print()
    print('Player2--Wins & Losses:')
    print('%s Wins, %s Losses, ' % (Player2Win, Player2Loss))
    print()
    print('No. of Draws:')
    print('%s Draws ' % (DrawCount))
    print()
    
    print('\n Play Again!!!\n\n')
    
```
